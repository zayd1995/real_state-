```python
def corr_chi2(data1, columns):
    for i in range(len(columns)):
        for j in range(i + 1, len(columns)): 
            Crosstab = pd.crosstab(data1[columns[i]], data1[columns[j]])
            stat, p, dof, expected = chi2_contingency(Crosstab)
            alpha = 0.05
            correlation = "correlated" if p < alpha else "not correlated"
            print(f"Columns {columns[i]} and {columns[j]} are {correlation} with p-value {p:.4f}")

columns = ['Town', 'Property Type', 'Residential Type']
corr_chi2(data1, columns)
```

```python
set1 = data1['Sale Amount']
set2 = data1['Assessed Value']

t_statistic, p_value = stats.ttest_rel(set1, set2)
alpha = 0.05
print(f"T-Statistic: {t_statistic}, P-Value: {p_value:.4f}")

if p_value < alpha:
        print(
            "Reject the null hypothesis. There is a statistically significant difference.")
else:
        print("Fail to reject the null hypothesis. There is no statistically significant difference.")
 
```

```python
set1 = data2['Sale Amount']
set2 = data2['Assessed Value']

t_statistic, p_value = stats.ttest_rel(set1, set2)
alpha = 0.05
print(f"T-Statistic: {t_statistic}, P-Value: {p_value:.4f}")

if p_value < alpha:
        print(
            "Reject the null hypothesis. There is a statistically significant difference.")
else:
        print("Fail to reject the null hypothesis. There is no statistically significant difference.")
```

```python
set1 = data1['Sale Amount']
set2 = data2['Sale Amount']

t_statistic, p_value = stats.ttest_ind(set1, set2)
alpha = 0.05
print(f"T-Statistic: {t_statistic}, P-Value: {p_value:.4f}")

if p_value < alpha:
        print(
            "Reject the null hypothesis. There is a statistically significant difference.")
else:
        print("Fail to reject the null hypothesis. There is no statistically significant difference.")
```
